```python
cols = ['date']
cols.extend(regions)
tiers = pd.DataFrame(columns=cols)

sd = start_date

while sd <= end_date:

    row = {}
    row['date'] = sd

    for r in changes.keys():
        if sd in changes[r]:
            row[r] = changes[r][sd]
        else:
            row[r] = '-'

    tiers = tiers.append(row, ignore_index=True)

    sd = sd + timedelta(days=1)
```
